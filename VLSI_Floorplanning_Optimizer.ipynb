```python
#Input---------->
capital = 1000
historical_prices = [-1.2, 3.4, -0.8, 2.1, -2.5, 1.7, -0.3, 5.8, -1.1, 3.5]
initial_population = [{"stop_loss": 2, "take_profit": 5, "trade_size": 20}, {"stop_loss": 3, "take_profit": 7,"trade_size": 30},{"stop_loss": 1.5, "take_profit": 4, "trade_size": 25},{"stop_loss": 2.5, "take_profit": 6, "trade_size": 15} ]
generations= 10
```

# Task-1

---

```python

import random

population = []

best_chromosome = []

#handle decimal point chromosome differently


def create_population(init_pop):

    population = []
    for chun in init_pop:

        var = ''
        for i, j in chun.items():
            if len(str(j)) == 1:
                var += '0'+str(int(j*10))
            elif len(str(j)) != 1 and '.' in str(j):
                var += '0'+str(int(j*10))
            else:
                var += str(int(j*10))
        population.append(var)

    return population


def fiteness_function(population,cap,hist_price):

    selection = [0,0]
    # print("popul----------->", population)
    for chromosome in population:

        stop_loss = None
        profit = None
        trade_size = None

        count = 0
        capital = cap
        best_score = 0 # Final - Initial


        #Converting the data
        if '0' == chromosome[0]:
            stop_loss = round(int(chromosome[1:3])/10,1)
        else:
            stop_loss = round(int(chromosome[0:3])/10,1)

        if '0' == chromosome[3]:
            profit =  round(int(chromosome[4:6])/10,1)
        else:
            profit =  round(int(chromosome[3:6])/10,1)

        if '0' == chromosome[6]:
            trade_size = int(int(chromosome[7:9])/10)
        else:
            trade_size = int(int(chromosome[6:9])/10)


        trade_capital = round((capital/100)*trade_size, 2)
        # print(f"Stop loss , Prof , Trade ")
        # print(stop_loss,profit,trade_size)


        #start of fitness score calculation->
        prof_loss = 0
        update = 0
        update_trade = 0
        exit_cond = "No SL/TP hit"

        fitness_score = []

        # print(f" Day | Price Change(%) | Trade size | Profit/Loss | Updated Capital ")
        for i in range(10):
            if hist_price[i]<0:
                if hist_price[i] <= stop_loss:
                    prof_loss = round(hist_price[i]*(trade_capital/100),2)
                else:
                    exit_cond = "TP hit"
                    prof_loss = round(stop_loss*(trade_capital/100),2)


                update = round(capital + prof_loss,2)

                capital = update
                # print("doooooooooooo",capital)

            else:
                if hist_price[i] <= profit:
                    prof_loss = round(hist_price[i]*(trade_capital/100),2)
                else:
                    exit_cond = "TP hit"
                    prof_loss = round(profit*(trade_capital/100) ,2)
                #price change == hist_price[i]

                update = round(capital + prof_loss,2)

                capital = update



            # print(f"Day->{i} Price Change(%)-> {hist_price[i]} Trade size-> {trade_capital} Profit/Loss {prof_loss} Updated Capital {capital} ")
            # print(f" {i} | {hist_price[i]} | {trade_capital} | {prof_loss} | {capital} ")

            trade_capital = round((capital/100)*trade_size, 2)
            fitness_score.append(capital)
            exit_cond = "No SL/TP hit"

        best = max(fitness_score)

        best_score = best-cap

        if best_score > selection[1]: #selecting the best chorosome
            selection = []
            selection.append(chromosome)
            selection.append(round(best_score,2))

        # print("----------------------")
        # print("selection",selection)

    return selection



# crossing over
def cross_over(popul):
    cross = []

    paren1, paren2 = random.sample(popul, 2)
    child1 = paren1[:len(paren1)//2] + paren2[len(paren2)//2:]
    child2 = paren2[:len(paren2)//2] + paren1[len(paren1)//2:]

    paren3, paren4 = random.sample(popul, 2)
    child3 = paren1[:len(paren1)//2] + paren2[len(paren2)//2:]
    child4 = paren2[:len(paren2)//2] + paren1[len(paren1)//2:]

    cross.append(child1)
    cross.append(child2)
    cross.append(child3)
    cross.append(child4)

    return cross


# mutation
def mutation(cross):

    mutate = cross
    if random.random() <= 0.05:
        mutate = []
        # print("================")
        for i in cross:
            chrom = i
            ind = random.randint(0,len(i)-2)
            n_gene = str(random.randint(0,9))
            mutate_ch = chrom[:ind] + n_gene + chrom[ind+1:]
            mutate.append(mutate_ch)

    return mutate



population = create_population(initial_population)

two_point_population = population #will be needed for task 2!

# print(population)


#Initiating-------->
best_chromosome = fiteness_function(population,capital,historical_prices) # running for the first time



#running for other gen---->


for i in range(generations-1):
    cross = cross_over(population)
    mutated = mutation(cross)
    get_chromosome = fiteness_function(mutated,capital,historical_prices)

    if get_chromosome[1]>best_chromosome[1]: #Updating If I get even better chromosome
        best_chromosome = get_chromosome

# print(best_chromosome)

print("Best Strategy:")

#Final Answer------->
final = best_chromosome[0]
sl = 0
pr = 0
ts =0
if '0' == final[0]:
    sl = round(int(final [1:3])/10,1)
else:
    sl = round(int(final[0:3])/10,1)

if '0' == final[3]:
    pr =  round(int(final[4:6])/10,1)
else:
    pr =  round(int(final[3:6])/10,1)

if '0' == final[6]:
    ts= int(int(final[7:9])/10)
else:
    ts = int(int(final[6:9])/10)

dic = {"stop_loss": sl, "take_profit": pr, "trade_size": ts}
print(dic)
print("Final_profit:", best_chromosome[1])


# for i in population:
#     print(len(i))
```
